```python
sb.countplot(x=df.JobRole)
plt.xlabel("Job Role")
plt.ylabel("Number of Employess")
plt.title("Distribution of Empolyees acoording to job role")
plt.xticks(rotation=90)
plt.show
```

### Distribution of Employees According to Job Role

1)Sales Executive has the highest number of employees
2)Research Scientist and Laboratory Technician also form a large portion
3)Human Resources has the lowest employee count
4)Managerial and Director roles have fewer employees, which is expected due to hierarchy

```python
sb.countplot(x=df.Department)
plt.xlabel("Department")
plt.ylabel("Number of Employess")
plt.title("Distribution of Empolyess acoording to Department")
plt.xticks(rotation=90)
plt.show
```

### Distribution of Employees According to Department
1)Job satisfaction is negatively correlated with attrition.
2)Salary shows a positive correlation with experience and job level.
3)Performance rating has a weak correlation with salary.
4)Experience-related variables are strongly correlated with each other.
5)Higher satisfaction and income are linked to lower employee turnover.

```python
sb.histplot(x=df.Age)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show
```

### Age Distribution
1)Majority of employees fall in the 25â€“40 years age group.

```python
sb.histplot(x=df.MonthlyIncome)
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.show
```

### Insights from the Monthly Income Histogram
1)Majority of employees earn between 2,000 and 6,000 per month.
2)The income distribution is right-skewed, with a small number of high earners.
3)Fewer employees fall in the higher income range (above 10,000).
4)This indicates a salary pyramid structure, with more lower-to-mid income employees and fewer high-income employees

```python
sb.countplot(x = df.Attrition)
plt.xlabel("Attrition Status")
plt.ylabel("Number Of Employees")
plt.title("Distribution of employees across Job Roles")
plt.xticks(rotation = 90)
plt.show()
```

### Insights from COUNT PLOT of HR Attrition Data
1)Majority of employees did not leave the organization.
2)A small proportion of employees have attrited compared to those retained.
3)The organization shows a relatively low attrition rate overall.
4)Retention appears stable, but the attrited group still requires analysis to identify risk factors.

```python
sb.boxplot(x = df.Attrition, y = df.JobSatisfaction)
```

### Insights from Job Satisfaction vs. Attrition
1)Employees who left the organization show lower job satisfaction levels.
2)Employees who stayed have a higher median job satisfaction.
3)Lower satisfaction is clearly associated with higher attrition.

```python
sb.boxplot(x = df.Attrition, y = df.MonthlyIncome)
```

### Insights from Monthly Income vs. Attrition
1)Employees who left the organization generally have lower monthly income.
2)Employees who stayed show a higher median income.
3)Lower income is associated with higher attrition

```python
sb.boxplot(x = df.Attrition, y = df.PerformanceRating)
```

### Insights from Performance Rating vs. Attrition
1)Performance ratings are similar for both attrited and retained employees.
2)Most employees have a performance rating around 3.
3)Performance rating does not strongly influence attrition.
